```python
#Hacer un plot con 4 plots mostrando cada uno una bandwidth diferente
bandwidths = [0.1, 0.2, 0.5, 1]

f = plt.figure()
f = f.set_size_inches(20,10)


for pos, bw in enumerate(bandwidths):
    plt.subplot(2, 2, pos + 1)   
    plt.title('Bandwith = %s' % bw)
    ax = sns.kdeplot(df['selling price'], bw = bw)
```

```python
#la duracion de los proximos 5 pases

def get_iss(lat, lon):
    
    url = 'http://api.open-notify.org/iss-pass.json?lat= %f&lon=%f' % (lat, lon) 
    
    response = requests.get(url)   
    iss = json.loads(response.content)    
    iss['response']
    
    
    
    return iss['response'] 
```

```python
get_iss(40,-3)
```

```python
def get_aflcio_alert():
    r = requests.get('http://www.aflcio.org/Legislation-and-Politics/Legislative-Alerts')
    soup = BeautifulSoup(r.content)
    result = []
    for alert in soup.find_all('div', class_='ec_statements'):
        dictionary = {}
        dictionary['date'] = alert.find(id = 'legalert_date').get_text()
        dictionary['title'] = alert.a.get_text()
        dictionary['link'] = 'http://www.aflcio.org' + alert.a['href']
        
        result.append(dictionary)
        
    return result
```

```python
letters = get_aflcio_alert()
json.dumps(letters)
```
